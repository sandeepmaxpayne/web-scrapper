# Sending a Post Request

```python
headers = {
    'Content-Type': 'application/json'
}
```

```python
recipie = {
    'title': 'Cappuccino',
    'ingr': ['18gm ground espresso', '150ml milk']
}
```

```python
r = requests.post(url, headers=headers, json=recipie)
r.status_code
```

```python
capp_info = r.json()
print(json.dumps(capp_info, indent=4))
```
