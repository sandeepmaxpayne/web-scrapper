# Simple Currency Converter

```python
date = input('Input date in format "yyyy-mm-dd" or the latest: ')
base = input('convert from (currency): ')
curr = input('convet to (currency): ')
quan = float(input('How much {} do you want to convert: '.format(base)))
url = base_url + "/" + date + "?base=" + base + "&symbols=" + curr
response = requests.get(url)

if (response.ok is False):
    print(f'\nError: {response.status_code}')
    print(response.json()['error'])
else:
    data = response.json()
    rate = data['rates'][curr]
    result = quan * rate
    
    print("\n {0} {1} is equals to {2} {3}, based upon the rate on {4}.".format(quan, base, result, curr, data['date']))
    
```
